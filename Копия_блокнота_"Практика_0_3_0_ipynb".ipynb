# **Задача 1.**




Типы данных Напишите программу на Python, которая принимает список чисел в качестве входных данных от пользователя, а затем выводит сумму, среднее значение и максимальное значение из списка.

***Нельзя использовать: len(), min(), max(), average(), sum()***

Дано:
```
1 2 3 4 5 6 7
```
Вывод:

```
Сумма чисел: 28
Среднее арифм.: 4.0
Максимальное значение: 7
```



````python
nums = []
while number := (input('Введите число:')):
  if number == '': pass
  nums.append(int(number))
min = 100000
maxi = -10**99
summa = 0
avg = 0
amount_num = 0
for i in range(0,len(nums)):
  if nums[i]<min: min = nums[i]
  if nums[i]>maxi: maxi = nums[i]
  summa += nums[i]
  amount_num += 1
avg = summa/amount_num
print('Cумма чисел:', summa)
print('Среднее арифм.:', avg)
print('Максимальное значение:', maxi)
````

Создайте простого чат-бота, который отвечает на основные запросы пользователей. Программа должна использовать условные выражения для определения намерений пользователя и последующего соответствующего ответа.

Чат бот может:

- помогать складывать числа

````python
from random import *

music = ['Chuck Berry - Johnny B. Goode', 'Jimi Hendrix - Purple Haze', 'AC/DC - Back in Black',
         'Nirvana - Smells Like Teen Spirit', 'Nirvana - Lithium', 'The Rolling Stones - Can’t You Hear Me Knocking',
         'Michael Jackson - Billie Jean', 'The Beatles - Yesterday', 'Oasis - Wonderwall', 'ABBA - Dancing Queen',
         'John Lennon - Imagine']
films = ['Побег из Шоушенка	(1994)', 'Крёстный отец (1972)', 'Тёмный рыцарь (2008)', 'Крёстный отец 2 (1974)',
         'Бойцовский клуб	(1999)', 'Форрест Гамп (1994)', 'Матрица (1999)', 'Молчание ягнят (1991)',
         'Назад в будущее (1985)', 'Титаник (1997)', 'Индиана Джонс: В поисках утраченного ковчега (1981)']
deals = ['']
nums = []
message_from_bot = ('''Я Чат-Бот Олег. Вот с чем я могу помочь:
1) Подсказать музыку
2) Подсказать фильм
3) Добавить заметку
4) Убрать заметку
5) Очистить список
6) Выполнить операцию сложения чисел
7) Прекратить взаимодействие с ботом
Выберите любую цифру: ''')

print(message_from_bot)

while operation_1 := input():
    if int(operation_1) not in range(1,8):
      print('Некорректный номер операции. Повтори ввод вновь:')
    if int(operation_1) == 1:
        print(f'Я вам посоветую: {music[(randint(0, len(music) - 1))]}')
        print(message_from_bot)

    if int(operation_1) == 2:
        print(f'Я вам посоветую: {films[(randint(0, len(films) - 1))]}')
        print(message_from_bot)

    if int(operation_1) == 3:
        deals.append(input('Добавьте вашу заметку: '))
        print(f'Заметка добавлена, ее номер {len(deals) - 1}.')
        print('Ваши дела:', deals[1::])
        print(message_from_bot)

    if int(operation_1) == 4:
        print(f'Ваш список заметок: {deals[1::]}')
        delete = deals.pop(int(input('Номер заметки, которую хотите удалить: ')))
        print(f'Вы удалили из списка дел заметку "{delete}" !')
        print('Ваши дела:', deals[1::])
        print(message_from_bot)

    if int(operation_1) == 5:
        print('Вы очистили список дел!')
        deals.clear()
        deals = ['']
        print(message_from_bot)

    if int(operation_1) == 6:
        nums.clear()
        while num := (
                input(
                    'Напишите свое число. Если вы ввели все необходимые ваши числа, то оставьте поле ввода пустым: ')):
            if num == '':
                pass
            else:
                nums.append(int(num))
        total = 0
        for num in nums:
            total += num
        print(f'Ваша сумма чисел: {total}')
        print(message_from_bot)

    if int(operation_1) == 7:
        print('Вы завершили работу с ботом. Всего хорошего!')
        break
````

# **Задача 4.**


Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Ввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
from random import *
items = ['камень','ножницы','бумага']
count_human = 0
count_pc = 0
round = 0
while (count_pc < 3) and (count_human < 3):
    round += 1
    print(f'Этап: {round}')
    print(f'Компьютер: {count_pc}, Игрок: {count_human}')
    human = input('Введите название предмета(камень,ножницы,бумага):').lower()
    computer = items[randint(0,2)]
    print('Компьютер:',computer)
    print('Игрок:',human)
    if human == computer:
      count_pc += 0
      count_human += 0
    if (human == items[0] and computer == items[1]) or (human == items[1] and computer == items[2]) or (human == items[2] and computer == items[0]):
      count_human += 1
    if (computer == items[0] and human == items[1]) or (computer == items[1] and human == items[2]) or (computer == items[2] and human == items[0]):
      count_pc += 1

if count_human == 3:
    print(f'Итоговый счет. Компьютер: {count_pc}, Игрок: {count_human}')
    print('Победа: Игрок')
if count_pc == 3:
    print(f'Итоговый счет. Компьютер: {count_pc}, Игрок: {count_human}')
    print('Победа: Компьютер')
````

# **Задача 5.**

Создайте игру "Палач" или "Виселица", в которой пользователь должен угадать слово, предлагая буквы. Программа должна выбрать случайное слово, а затем разрешить пользователю угадывать буквы. После каждого угадывания программа должна сообщать пользователю, есть ли в слове буква или нет.

**Этап 1**
Создать саму игру

**Этап 2:**
Ввести подсчет количества очков и ограничения на попытки

**Этап 3:** Добавить возможность сразу угадать слово

````python
from random import choice


words = ['гренки','болт','яблоко', 'дерево', 'кошка', 'собака',\
         'математика', 'компьютер', 'телефон','ручка', \
         'лампочка', 'диван', 'ваза', 'апостол', 'передряга', 'бойня', \
         'въезд', 'выезд', 'проезд', 'автобус', 'программирование', 'филология', 'курица', \
         'булка','кисточка', 'миньон', 'квартира', 'роутер', 'алкоголь',\
         'мышка', 'монитор', 'аукцион', 'изморозь', 'гололёд', 'лампа', \
         'картина', 'военкомат', 'университет', 'радиофак', 'бухгалтер']
win_word = choice(words)
hide_word = ['_' for i in range(len(win_word))]
mistakes = 0
max_attempts = int(input('Введите кол-во ваших попыток: '))
used_letters = []
score = 0
flag = 1

print('Загаданное слово: ', *hide_word)
user_way = int(input("""
Угадать слово - 1 (при неудачной попытке проигрыш, при удачной +300 очков)
Выбрать букву - 2 (при неудачной попытке -5 баллов, -1 право на ошибку, за каждую угаданную букву +15 баллов
Выберите стратегию: """))

while (flag != 0) and(max_attempts > 0) and (win_word != ''.join(hide_word)):
    mistakes = 0
    print('Загаданное слово: ', *hide_word)

    while user_way == 2:
        letter = input('Введите вашу букву: ')
        while letter in used_letters:
            letter = input('Вы уже использовали эту букву: Введите вашу букву: ')

        if win_word == ''.join(hide_word):
            print(f'Поздравляем, вы выиграли!')
            print('Загаданное слово: ', ''.join(hide_word))
            print('Очки: ', score)
            exit()

        if letter in win_word:
            used_letters.append(letter)
            for i in range(0,len(hide_word)):
                if letter == win_word[i]:
                    score += 15
                    print(f'Вы угадали букву! +15 очков, очки: {score}')
                    hide_word[i] = letter
                    if win_word == ''.join(hide_word):
                        print(f'Поздравляем, вы выиграли! Загаданное слово {win_word}, очки: {score}')
                        exit()
            user_way = int(input(f"""
Оставшееся кол-во попыток: {max_attempts-mistakes}
Загаданное слово: {hide_word}
Угадать слово - 1 (при неудачной попытке проигрыш, при удачной +300 очков)
Выбрать букву - 2 (при неудачной попытке -5 баллов, -1 право на ошибку, за каждую угаданную букву +15 баллов
Ваша буква входит в слово! Повторно выберите стратегию: """))

        if letter not in win_word:
            print('Загаданное слово: ', *hide_word)
            used_letters.append(letter)

            mistakes += 1
            if mistakes == max_attempts:
                print('Количество попыток исчерпано. вы повешаны')
                print(f'Загаданное слово: {win_word}')
                exit()
            user_way = int(input(f"""
Оставшееся кол-во попыток: {(max_attempts-mistakes)}
Загаданное слово: {hide_word}
Угадать слово - 1 (при неудачной попытке проигрыш, при удачной +300 очков)
Выбрать букву - 2 (при неудачной попытке -5 баллов, -1 право на ошибку, за каждую угаданную букву +15 баллов
К сожалению, такой буквы в слове нет, -5 очков. Попробуйте вновь выбрать стратегию: """))
            score -= 5

    while user_way == 1:
        print('Загаданное слово: ', *hide_word)
        user_word = input('Введите ваше слово: ')
        if user_word == ''.join(win_word):
            score += 300
            print(f'Поздравляем, вы выиграли! Загаданное слово: {win_word}, очки: {score}')
            exit()

        if user_word != ''.join(win_word):
            print(f'Вы были повешены. Сожалеем. Загаданное слово: {win_word}, очки: {score}')
            flag = 0
            exit()

````

# **Дополнительно: Задача 6.**



**Цифра на определенном месте:**

Последовательно записан натуральный ряд чисел.

Какая цифра стоит в N позиции


Вввод:

```
Введите номер позиции: 1234567890
```

Вывод:

```
8
```

````python
length = int(input('Введите длину(кол-во чисел) последовательности натуральных чисел: '))
numbers = '' #строка, над которой производятся операции для вывода её пользователю
result = '' #строка идентичная numbers, по которой происходит поиск цифры
count = 0

for i in range(1,length+1):
  numbers += str(i)
  numbers += ' '
  result += str(i)

print('Ваша последовательность натуральных чисел:', numbers) # последовательность
print('Всего цифр в последовательности:',len(result))
N = int(input('Введите позицию цифры(не учитывая пробелы между числами), которую хотите вывести:'))

while N > len(result):
  print('Вы ввели порядковый номер цифры, который превышает их кол-во. Повторите ввод вновь.')
  N = int(input('Введите позицию цифры(не учитывая пробелы между числами), которую хотите вывести:'))

print(f'Цифра, стоящая на {N} позиции:', result[N-1])
````
